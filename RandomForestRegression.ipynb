### DECISION TREE REGRESSION 

```python
## IMPORT THE LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests,io
```

```python
## import the datasets
url = "https://drive.google.com/u/0/uc?id=1mVmGNx6cbfvRHC_DvF12ZL3wGLSHD9f_&export=download"
data = requests.get(url).content
```

```python
## reading the data
dataset = pd.read_csv(io.StringIO(data.decode('utf-8')))
dataset.head()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].scatter(dataset["Petrol_tax"], dataset["Petrol_Consumption"], color='b')
axes[0, 0].set_xlabel("Petrol Tax")
axes[0, 0].set_ylabel("Petrol Consumption")
axes[0, 0].set_title("Petrol Tax vs Petrol Consumption")

axes[0, 1].scatter(dataset["Average_income"], dataset["Petrol_Consumption"], color='r')
axes[0, 1].set_xlabel("Average Income")
axes[0, 1].set_ylabel("Petrol Consumption")
axes[0, 1].set_title("Average Income vs Petrol Consumption")

axes[1, 0].scatter(dataset["Paved_Highways"], dataset["Petrol_Consumption"], color='g')
axes[1, 0].set_xlabel("Paved Highways")
axes[1, 0].set_ylabel("Petrol Consumption")
axes[1, 0].set_title("Paved Highways vs Petrol Consumption")

axes[1, 1].scatter(dataset["Population_Driver_licence(%)"], dataset["Petrol_Consumption"], color='purple')
axes[1, 1].set_xlabel("Driver Licence (%)")
axes[1, 1].set_ylabel("Petrol Consumption")
axes[1, 1].set_title("Driver Licence % vs Petrol Consumption")

plt.tight_layout()
plt.show()
```

## seperate the features and the target variable

```python
x = dataset.drop('Petrol_Consumption', axis= 1) #features
y = dataset['Petrol_Consumption'] #target
```

```python
## validate model
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=10, random_state=0)
model.fit(x,y)
```

```python
y_pred = model.predict([[8.0,4599,541,0.924]])
y_pred
```
